1. What is NumPy, and why is it widely used in Python?

ANS:- NumPy is a Python library used for working with large arrays and mathematical operations. It is widely used because it is very fast, memory-efficient, and supports powerful functions for scientific and numerical computing.

2. How does broadcasting work in NumPy?

ANS:- Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array so that both arrays have compatible shapes.

3. What is a Pandas DataFrame?

ANS:- A Pandas DataFrame is a two-dimensional data structure similar to a table with rows and columns. It is used to store and manipulate structured data easily.

4. Explain the use of the groupby() method in Pandas?

ANS:- The groupby() method is used to group data based on a column and then perform operations like sum, mean, or count on each group.

5. Why is Seaborn preferred for statistical visualizations?

ANS:- Seaborn is preferred because it provides attractive default styles and makes it easy to create complex statistical plots with less code.

6. What are the differences between NumPy arrays and Python lists?

ANS:- NumPy arrays are faster, use less memory, and support mathematical operations directly, while Python lists are more flexible but slower for numerical tasks.

ANS:- A heatmap is a graphical representation of data using colors. It is used to show correlations or intensity values in matrices.

8. What does the term “vectorized operation” mean in NumPy?

ANS:- Vectorized operations allow calculations on entire arrays without using loops, making them faster and more efficient.

```python
# 1.  How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# 2.Write a Pandas script to find the mean of a specific column in a DataFrame?
import pandas as pd

data = {'Marks': [80, 90, 70]}
df = pd.DataFrame(data)
print(df['Marks'].mean())
```

```python
# 3.  Create a scatter plot using Matplotlib?
import matplotlib.pyplot as plt

plt.scatter([1,2,3], [4,5,6])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

corr_matrix = df.select_dtypes(include='number').corr()

sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
# 5.Generate a bar plot using Plotly?
import plotly.express as px

data = px.data.tips()
fig = px.bar(data, x="day", y="total_bill")
fig.show()
```

```python
# 6.Create a DataFrame and add a new column based on an existing column?
import pandas as pd

data = {
    'Marks': [60, 75, 90, 50]
}

df = pd.DataFrame(data)

df['Bonus Marks'] = df['Marks'] + 5

print(df)
```

```python
# 9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
import pandas as pd

data = {
    'Name': ['Rahul', 'Anita', 'John', 'Neha'],
    'Marks': [65, 85, 55, 90]
}

df = pd.DataFrame(data)

threshold = 70

filtered_df = df[df['Marks'] > threshold]

print(filtered_df)
```

```python
# 10.Create a histogram using Seaborn to visualize a distribution?
import seaborn as sns
import matplotlib.pyplot as plt
data = [10, 20, 20, 30, 30, 30, 40, 50, 60, 60, 70]
sns.histplot(data, bins=5)

plt.title("Histogram of Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# 11.  Perform matrix multiplication using NumPy?
import numpy as np

matrix1 = np.array([[1, 2],
                    [3, 4]])

matrix2 = np.array([[5, 6],
                    [7, 8]])
result = np.dot(matrix1, matrix2)

print(result)
```

```python
# 13.  Create a 3D scatter plot using Plotly?
import plotly.express as px
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 70, 80, 90],
    'Age': [15, 18, 20, 22, 25]
}

fig = px.scatter_3d(
    data,
    x='Height',
    y='Weight',
    z='Age',
    title='3D Scatter Plot of Students'
)
fig.show()
```
